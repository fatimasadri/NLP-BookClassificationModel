```python
import requests
import pandas as pd
import time

titles = []
page = 1

while len(titles) < 750 and page <= 50:
    url = "https://openlibrary.org/search.json"
    params = {
        "subject": "fiction",
        "language": "eng",
        "page": page
    }
    r = requests.get(url, params=params, timeout=30)
    data = r.json()

    for doc in data.get("docs", []):
        t = doc.get("title")
        if t and len(t) > 1:
            titles.append(t.strip())

    page += 1
    time.sleep(0.2)  # polite delay

# Deduplicate and cut to 750
titles = list(dict.fromkeys(titles))[:750]

df = pd.DataFrame({"book_title": titles, "genre": ["Fiction"] * len(titles)})
df.to_csv("fiction_750_openlibrary.csv", index=False)

print(df.head())
print("Saved:", len(df), "rows")
```
